```python
import pint.models.model_builder as mb
import pint.toa as toa
import libstempo as lt
import matplotlib.pyplot as plt
%matplotlib inline
import tempo2_utils
import astropy.units as u
from pint.residuals import resids
import numpy as np
from astropy.table import Table, Column
import pint.utils as ut
```

# Set up model for B1855

```python
parfile = 'B1855+09_NANOGrav_dfg+12_modified.par'
timfile = 'B1855+09_NANOGrav_dfg+12.tim'
psr = lt.tempopulsar(parfile, timfile)
mdd = mb.get_model(parfile)
```

```python
toas = toa.get_TOAs(timfile,ephem="DE405", planets=False)
```

# Get tempo2 general2 plugin results

```python
tempo2_vals = tempo2_utils.general2(parfile, timfile,
                                    ['tt2tb', 'roemer', 'post_phase','tt',
                                     'shapiro', 'shapiroJ','bat','clock0','clock1','clock2','clock3','clock4','sat','tropo'
                                     ,'shapiroS','ism','clock','shapiroN','shapiroV','shapiroU'])
```

# Plot the residuals for PINT

```python
mjds = toas.get_mjds()
mjds = np.sort(mjds)
```

```python
tt = toas.table
```

```python
resids_us = resids(toas, mdd).time_resids.to(u.us)
plt.plot(mjds,resids_us)
```

# Plot residuals difference PINT-TEMPO2

```python
diff = resids_us - (psr.residuals()*u.s).to(u.us)
plt.plot(mjds,diff-diff.mean())
plt.ylabel('PINT-T2(us)')
```

# With gps2utc clock correction

```python
diff = resids_us - (psr.residuals()*u.s).to(u.us)+tempo2_vals['clock1']*u.s
plt.plot(mjds,diff-diff.mean())
plt.ylabel('PINT-T2(us)')
```

# Test the shaprio delay for planets. Here is the problem

```python
def planet_shapiro(model,toaTable):
    delay = {}
    psr_dir = model.ssb_to_psb_xyz(epoch=toaTable['tdbld'].astype(np.float64))
    for pl in ('jupiter', 'saturn', 'venus', 'uranus'):
        delay[pl] = model.ss_obj_shapiro_delay(toaTable['obs_'+pl+'_pos'],
                                                   psr_dir, model._ss_mass_sec[pl])
    return delay
```

# Get PINT shapiro delay calculation

```python
sDelay = planet_shapiro(mdd,tt)
```

# Jupiter Shapiro delay difference

```python
plt.plot(mjds,sDelay['jupiter']-tempo2_vals['shapiroJ'])
```

# It is quite big. Let us look the shapiro delay individually

```python
plt.plot(mjds,sDelay['jupiter'])
```

```python
plt.plot(mjds,tempo2_vals['shapiroJ'])
```

```python
plt.plot(mjds,sDelay['jupiter']-tempo2_vals['shapiroJ'])
```

# Test for J1744-1134

```python
par2 = 'J1744-1134.basic.par'
tim2 = 'J1744-1134.Rcvr1_2.GASP.8y.x.tim'
psr2 = lt.tempopulsar(par2, tim2)
mi = mb.get_model(par2)
```

```python
toas2 = toa.get_TOAs(tim2,ephem="DE421",planets=True)
```

```python
tt2 = toas2.table
```

```python
tempo2_vals2 = tempo2_utils.general2(par2, tim2,
                                    ['tt2tb', 'roemer', 'post_phase','tt',
                                     'shapiro', 'shapiroJ','bat','clock0','clock1','clock2','clock3','clock4','sat','tropo'
                                     ,'shapiroS','ism','clock','earth_ssb1','earth_ssb2','earth_ssb3','shapiroN','shapiroU',
                                     'shapiroV'])
```

```python
resids_us2 = resids(toas2, mi).time_resids.to(u.us)
```

```python
t2res = psr2.residuals()
```

# Residuals difference for J1744-1134

```python
mv2 = (resids_us2-(t2res*u.s).to(u.us)).mean()
plt.plot(resids_us2-(t2res*u.s).to(u.us)-mv2)
plt.ylabel('PINT-TEMPO2(us)')
```

# With gps2utc clock correction

```python
mv2 = (resids_us2-(t2res*u.s).to(u.us)+tempo2_vals2['clock1']*u.s).mean()
plt.plot(resids_us2-(t2res*u.s).to(u.us)+tempo2_vals2['clock1']*u.s-mv2)
plt.ylabel('PINT-TEMPO2(us)')
```

# Get the planet shapiro delay

```python
delay2 = planet_shapiro(mi,tt2)
```

# Jupiter shapiro delay difference PINT-TEMPO2

```python
plt.plot(delay2['jupiter']-tempo2_vals2['shapiroJ'])
plt.ylabel('Shapiro delay difference (s)')
```

# It is the same level with B1855. Let us look individually

```python
plt.plot(delay2['jupiter'])
plt.title('PINT jupiter shapirodelay')
```

```python
plt.plot(tempo2_vals2['shapiroJ'])
plt.title('tempo2 jupiter shapirodelay')
```
